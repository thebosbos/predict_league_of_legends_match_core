```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('games.csv')
df
```

# Nouvelle section

```python
x=df.iloc[:,5:]

y=df.iloc[:,4]
```

```python
data=df[["winner","firstBlood","firstTower","firstInhibitor","firstBaron","firstDragon","firstRiftHerald","t1_towerKills","t1_inhibitorKills",'t1_baronKills',
       't1_dragonKills','t2_towerKills','t2_inhibitorKills', 't2_baronKills', 't2_dragonKills']]
data=data.fillna(data.mean())
```

```python
x=data.iloc[:,1:].values
y=data.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,classification_report
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc= RandomForestClassifier(n_estimators=200,#number of tree
                            criterion='entropy',
                            random_state=50,#minimum number of samples required to split an internal node
                            min_samples_split=5,#lena el separation fel arbre 
                            min_samples_leaf=3,#minimum number of samples required to be at a leaf noce.
                            max_features="log2",#max caracteriltique fi kol arbre ma3neha el caracteristique fi kol arbre = lenght mta3 les cacarteristiqye / log2
                            oob_score=True,#use out-of-bag samples to estimate the generalization score
                            class_weight="balanced",
                            n_jobs=10,)
```

```python
rfc.fit(xtrain,ytrain)
```

```python
predicted_output = rfc.predict(xtest)
```

```python
print("***************** Performance evaluation of RF*****************************")
print(confusion_matrix(ytest,predicted_output)) #lena i9aren bin el prediction eli 3malha w les resiltat mta3 el expret
print(classification_report(ytest,predicted_output))
```
